```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('diabetes.csv')
```

```python

# Select numerical columns for outlier detection
numerical_cols = df.select_dtypes(include=['float64', 'int64']).columns

# Define a function to trim outliers using the IQR method
def trim_outliers(df, columns):
    for col in columns:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        df = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]
    return df


# Trim outliers from the numerical columns
df_trimmed = trim_outliers(df.copy(), numerical_cols)

```

```python
sns.boxplot(data=df_trimmed)
```

```python
# Define a function to cap outliers using the IQR method
def cap_outliers(df, columns):
    for col in columns:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        df[col] = df[col].apply(lambda x: lower_bound if x < lower_bound else (upper_bound if x > upper_bound else x))
    return df
```

```python
# Cap outliers in the numerical columns
df_capped = cap_outliers(df.copy(), numerical_cols)

```

```python
sns.boxplot(data=df_capped)
```
